```python
# Image attributes
img_path='fish.JPEG'

# img_path='car.JPEG'
img = Image.open(img_path).convert('RGB')
image = np.array(img)
image = np.transpose(image,(1,0,2))
image = np.transpose(image,[1,0,2])
print(image.shape)
height=image.shape[0]
width=image.shape[1]
imshow(image)
```
